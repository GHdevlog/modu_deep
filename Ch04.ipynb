# 최소 제곱법

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

# 평균값 계산
mx = np.mean(x)
my = np.mean(y)

# 분자 계산 함수
def top(x, mx, y, my):
    d = 0
    for i in range(len(x)):
        d += (x[i] - mx) * (y[i] - my)
    return d

# 분자와 분모 계산
dividend = top(x, mx, y, my)
divisor = sum([(i - mx) ** 2 for i in x])

# 기울기와 y절편 계산
a = dividend / divisor
b = my - (mx * a)

# 결과 출력
print(f"x의 평균값: {mx}")
print(f"y의 평균값: {my}")
print()
print(f"분자: {dividend}")
print(f"분모: {divisor}")
print()
print(f"기울기 a: {a}")
print(f"y 절편: {b}")

# 시각화
plt.scatter(x, y, color='blue', label='Data Points')  # 데이터 포인트 시각화
plt.plot(x, a*x + b, color='red', label=f'Linear Regression (y = {a:.2f}x + {b:.2f})')  # 회귀선 시각화

# 그래프 제목 및 축 레이블
plt.title('Linear Regression')
plt.xlabel('x')
plt.ylabel('y')

# 범례 추가
plt.legend()

# 그래프 출력
plt.show()
```

# 평균 제곱 오차(MSE)

```python
import numpy as np

# 가중치와 절편 설정 (임의로 설정된 값)
fake_a = 2.3  # 기울기
fake_b = 79  # y 절편

# 데이터: 공부 시간 (x)와 실제 점수 (y)
x = np.array([2, 4, 6, 8])  # 공부 시간
y = np.array([81, 93, 91, 97])  # 실제 점수

# 예측 함수: 공부 시간에 따라 점수를 예측
def predict(x):
    return fake_a * x + fake_b  # y = ax + b (회귀식)

predict_result = []  # 예측 결과를 저장할 리스트

# 각 데이터에 대해 예측 점수를 계산하고 출력
for i in range(len(x)):
    predict_result.append(predict(x[i]))  # 예측 점수를 리스트에 추가
    # 공부 시간, 실제 점수, 예측 점수를 출력
    print("공부시간=%.f, 실제점수=%.f, 예측점수=%.f" % (x[i], y[i], predict(x[i])))

n = len(x)  # 데이터 개수

# 평균 제곱 오차(MSE) 함수 정의
def mse(y, y_pred):
    return (1/n) * sum((y - y_pred)**2)  # MSE 계산: 실제값과 예측값의 차이의 제곱의 평균

# 평균 제곱 오차 출력
print("평균 제곱 오차: " + str(mse(y, predict_result)))

# 시각화
plt.scatter(x, y, color='blue', label='Actual Score')  # 실제 점수 산포도
plt.plot(x, predict_result, color='red', label='Predicted Score')  # 예측 선 그래프

# 그래프 제목과 축 라벨 추가 (영어로 변경)
plt.title('Actual Score vs Predicted Score')
plt.xlabel('Study Time (hours)')
plt.ylabel('Score')

# 범례 추가
plt.legend()

# 그래프 출력
plt.show()
```

# 경사하강법(Gradient Descent)

## 단순 선형 회귀

```python
import numpy as np  # numpy 라이브러리 임포트
import matplotlib.pyplot as plt  # matplotlib의 pyplot 모듈 임포트

# 공부 시간 (x)과 실제 점수 (y) 데이터를 배열로 생성
x = np.array([2, 4, 6, 8])  # 공부 시간
y = np.array([81, 93, 91, 97])  # 실제 점수

# 공부 시간과 실제 점수를 산점도로 표시
plt.scatter(x, y)
plt.show()

# 기울기(a)와 절편(b)를 초기화 (0으로 설정)
a = 0
b = 0

# 학습률 설정
learning_rate  = 0.01

# 반복 횟수 설정 (epochs)
epochs = 5001

# 데이터의 개수 (n)
n = len(x)

# 경사 하강법 알고리즘을 통해 최적의 기울기와 절편을 찾는 과정
for i in range(epochs):
    # 예측 값 계산 (현재 기울기와 절편을 사용하여)
    y_pred = a * x + b
    
    # 실제 값과 예측 값의 차이 (오차)
    error = y - y_pred
    
    # 기울기에 대한 편미분 (오차를 줄이기 위한 기울기의 변화량)
    a_diff = (2/n) * sum(-x * error)
    
    # 절편에 대한 편미분 (오차를 줄이기 위한 절편의 변화량)
    b_diff = (2/n) * sum(-error)
    
    # 기울기(a)와 절편(b)을 학습률을 곱한 변화량만큼 업데이트
    a = a - learning_rate  * a_diff
    b = b - learning_rate  * b_diff
    
    # 50회마다 현재의 기울기와 절편을 출력
    if i % 50 == 0:
        print("epochs=%.f, 기울기=%.04f, 절편=%.04f" % (i, a, b))

# 최종적으로 구한 기울기와 절편을 사용하여 예측 값을 계산
y_pred = a * x + b

# 실제 점수와 예측 값을 산점도와 직선으로 표시
plt.scatter(x, y)  # 실제 점수
plt.plot(x, y_pred, 'r')  # 예측 직선 (붉은색)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Study hours (x) and actual scores (y) data
x = np.array([2, 4, 6, 8])  # Study hours
y = np.array([81, 93, 91, 97])  # Actual scores

# MSE calculation function
def mse(a, b, x, y):
    y_pred = a * x + b
    return np.mean((y - y_pred) ** 2)

# Define ranges for a and b
a_vals = np.linspace(0, 5, 100)
b_vals = np.linspace(0, 150, 100)

# Create meshgrid for 3D graph
A, B = np.meshgrid(a_vals, b_vals)

# Initialize MSE values
MSE = np.zeros_like(A)

# Calculate MSE for each combination of a and b
for i in range(len(a_vals)):
    for j in range(len(b_vals)):
        MSE[i, j] = mse(a_vals[i], b_vals[j], x, y)

# Plot 3D graph
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot surface
ax.plot_surface(A, B, MSE, cmap='viridis')

# Set labels
ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_zlabel('MSE')

plt.title('MSE(a, b)')
plt.show()
```

## 다중 선형 회귀

```python
import numpy as np  # numpy 라이브러리 임포트
import matplotlib.pyplot as plt  # matplotlib의 pyplot 모듈 임포트
from mpl_toolkits.mplot3d import Axes3D  # 3D 그래프를 위한 모듈 임포트

# 공부 시간 (x1)과 과외 시간 (x2), 실제 점수 (y) 데이터를 배열로 생성
x1 = np.array([2, 4, 6, 8])  # 공부 시간
x2 = np.array([0, 4, 2, 3])  # 과외 시간
y = np.array([81, 93, 91, 97])  # 실제 점수

# 초기 기울기와 절편 설정
a1 = 0
a2 = 0
b = 0

# 학습률 설정
learning_rate  = 0.01

# 반복 횟수 설정 (epochs)
epochs = 10001

# 데이터의 개수 (n)
n = len(x1)

# 경사 하강법 알고리즘을 통해 최적의 기울기와 절편을 찾는 과정
for i in range(epochs):
    # 예측 값 계산 (현재 기울기와 절편을 사용하여)
    y_pred = a1*x1 + a2*x2 + b
    
    # 실제 값과 예측 값의 차이 (오차)
    error = y - y_pred
    
    # 기울기에 대한 편미분 (오차를 줄이기 위한 기울기의 변화량)
    a1_diff = (2/n) * sum(-x1 * error)
    a2_diff = (2/n) * sum(-x2 * error)
    
    # 절편에 대한 편미분 (오차를 줄이기 위한 절편의 변화량)
    b_diff = (2/n) * sum(-error)
    
    # 기울기(a)와 절편(b)을 학습률을 곱한 변화량만큼 업데이트
    a1 = a1 - learning_rate  * a1_diff
    a2 = a2 - learning_rate  * a2_diff
    b = b - learning_rate  * b_diff
    
    # 100회마다 현재의 기울기와 절편을 출력
    if i % 100 == 0:
        print("epochs=%.f, 기울기1=%.04f, 기울기2=%.04f, 절편=%.04f" % (i, a1, a2, b))

# 최종적으로 구한 기울기와 절편을 사용하여 예측 값을 계산
y_pred = a1*x1 + a2*x2 + b

print("실제 점수: ", y)
print("예측 점수: ", y_pred)

# 3D 그래프 그리기
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 실제 점수를 3D 산점도로 표시
ax.scatter(x1, x2, y, color='blue', label='Actual Scores')  # 실제 점수 (파란색 점)

# 예측 평면을 표시하기 위한 매쉬그리드 생성
x1_surf, x2_surf = np.meshgrid(np.linspace(min(x1), max(x1), 100), np.linspace(min(x2), max(x2), 100))

# 예측 평면의 z값 (y_pred 계산)
y_pred_surf = a1 * x1_surf + a2 * x2_surf + b

# 예측 평면을 3D 그래프에 표시
ax.plot_surface(x1_surf, x2_surf, y_pred_surf, color='red', alpha=0.5, rstride=100, cstride=100)

# 축 레이블 설정
ax.set_xlabel('Study Hours (x1)')
ax.set_ylabel('Tutoring Hours (x2)')
ax.set_zlabel('Scores (y)')

# 범례 표시
ax.legend()

# 그래프 타이틀
plt.title('Actual Scores and Predicted Plane')

# 그래프 표시
plt.show()
```

# 선형회귀 (With Tensorflow)

## 단순 선형 회귀

```python
import numpy as np
import matplotlib.pyplot as plt

# TensorFlow Keras에서 필요한 모듈을 임포트합니다.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 입력 데이터 (공부 시간)
x = np.array([2, 4, 6, 8])  # 공부 시간

# 출력 데이터 (시험 점수)
y = np.array([81, 93, 91, 97])  # 실제 점수

# Sequential 모델을 생성합니다.
model = Sequential()

# 첫 번째(그리고 유일한) 레이어로 Dense 레이어를 추가합니다.
# 입력 차원은 1, 출력 차원은 1이며, 활성화 함수는 'linear'로 설정합니다.
model.add(Dense(1, input_dim=1, activation='linear'))

# 모델을 컴파일합니다. 옵티마이저는 SGD(확률적 경사 하강법)를 사용하고,
# 손실 함수는 MSE(평균 제곱 오차)로 설정합니다.
model.compile(optimizer='sgd', loss='mse')

# 모델을 학습시킵니다. 주어진 입력(x)과 출력(y) 데이터를 사용하여
# 2000번의 epoch 동안 학습을 진행합니다.
model.fit(x, y, epochs=2000)

# 입력 데이터와 실제 데이터를 산점도로 시각화합니다.
plt.scatter(x, y)

# 모델이 예측한 값을 선 그래프로 시각화합니다.
plt.plot(x, model.predict(x), 'r')

# 그래프를 보여줍니다.
plt.show()
```

```python
import numpy as np

hour = 7
prediction = model.predict(np.array([hour]))
print(prediction)

print("%.f시간을 공부할 경우의 예상 점수는 %.02f점 입니다." % (hour, prediction[0][0]))
```

## 다중 선형 회귀

```python
import numpy as np
import matplotlib.pyplot as plt

# TensorFlow Keras에서 필요한 모듈을 임포트합니다.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 입력 데이터 (공부 시간, 과외 시간)
x = np.array([[2,0], [4,4], [6,2], [8,3]])

# 출력 데이터 (시험 점수)
y = np.array([81, 93, 91, 97]) 

# Sequential 모델을 생성합니다.
model = Sequential()

# 첫 번째(그리고 유일한) 레이어로 Dense 레이어를 추가합니다.
# 입력 차원은 1, 출력 차원은 1이며, 활성화 함수는 'linear'로 설정합니다.
model.add(Dense(1, input_dim=2, activation='linear'))

# 모델을 컴파일합니다. 옵티마이저는 SGD(확률적 경사 하강법)를 사용하고,
# 손실 함수는 MSE(평균 제곱 오차)로 설정합니다.
model.compile(optimizer='sgd', loss='mse')

# 모델을 학습시킵니다. 주어진 입력(x)과 출력(y) 데이터를 사용하여
# 2000번의 epoch 동안 학습을 진행합니다.
model.fit(x, y, epochs=2000)
```

```python
import numpy as np

hour = 7
private_class = 4
prediction = model.predict(np.array([[hour, private_class]]))
print(prediction)

print("%.f시간을 공부하고 %.f시간의 과외를 받을 경우의 예상 점수는 %.02f점 입니다." % (hour, private_class, prediction[0][0]))
```
